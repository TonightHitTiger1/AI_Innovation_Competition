```python
import pandas as pd
import numpy as np
```

```python
tea = pd.read_csv('result-teacher.csv')
stu = pd.read_csv('result-student.csv')
```

```python
tea = tea.reindex()
tea = tea.drop(['Unnamed: 0'],axis=1)
tea = tea.drop(['Unnamed: 0.1'],axis=1)

stu = stu.reindex()
stu = stu.drop(['Unnamed: 0'],axis=1)
stu = stu.drop(['Unnamed: 0.1'],axis=1)
```

```python
# tea
tea.dropna(axis=0,how='any',inplace=True)
tea
```

```python
stu.groupby(['person_class'])['person_class'].count()
```

```python
#  替换对应数字
# student = 0 
# student-back = 1
# teacher = 2
# teacher-back = 3
# teacher-write = 4
# student-stand = 5

# tea.columns

tea.loc[tea['person_class'] == 'student','person_class'] = 0
tea.loc[tea['person_class'] == 'student-back','person_class'] = 1
tea.loc[tea['person_class'] == 'teacher','person_class'] = 2
tea.loc[tea['person_class'] == 'teacher-back','person_class'] = 3
tea.loc[tea['person_class'] == 'teacher-write','person_class'] = 4

stu.loc[stu['person_class'] == 'student','person_class'] = 0
stu.loc[stu['person_class'] == 'student-back','person_class'] = 1
stu.loc[stu['person_class'] == 'student-stand','person_class'] = 5


```

```python
# 将学生csv与教师csv合并  
all_data = pd.concat([tea,stu])
all_data.reindex()
```

```python
target = all_data.pop('person_class')
target
```

```python
# 建立svm分类器
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score,roc_auc_score
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import lightgbm as lgb

# data split
X_train,X_test,y_train,y_test=train_test_split(all_data,list(target),test_size=1/5)
```

```python
# 随机森林
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
pre_test = rf.predict(X_test)
accuracy = accuracy_score(y_test,pre_test)
print("随机森林 accuracy:",accuracy)
```

```python
# svm分类器
clf = svm.SVC(kernel='linear',C=100)
clf.fit(X_train,y_train)
a=clf.predict(X_test)
b=clf.get_params() #查看模型的参数
c=clf.score(X_test,y_test)#查看匹配度，正确率
print("svm = ",c)
```

```python
# xgboost

model = XGBClassifier()               # 载入模型
model.fit(X_train,y_train)            # 训练模型（训练集）
y_pred = model.predict(X_test)        # 模型预测（测试集）
accuracy = accuracy_score(y_test,y_pred)
print("xgb accuarcy: %.2f%%" % (accuracy*100.0))
```
